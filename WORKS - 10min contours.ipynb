```python
# Central location (Longitude, Latitude)
center = [-3.6662352, 40.3871829]
```

```python
# Isochrone API endpoint
url = "https://api.mapbox.com/isochrone/v1/mapbox/walking/" + f"{center[0]},{center[1]}"

# Request parameters
```

```python
# Request parameters
params = {
    'contours_minutes': '10',  # 10 minutes walking
    'polygons': 'true',
    'access_token': mapbox_access_token
}
```

```python
# Send GET request to Mapbox Isochrone API
response = requests.get(url, params=params)
```

```python
# Get the JSON response body
data = response.json()
```

```python
# Get the isochrone as a Shapely polygon
polygon = shape(data['features'][0]['geometry'])
```

```python
# Create a list to store the routes
allroutes = []
```

```python
# Loop over each point in the polygon
for point in polygon.exterior.coords:
    # Directions API endpoint
    url = "https://api.mapbox.com/directions/v5/mapbox/walking/" + f"{center[0]},{center[1]};{point[0]},{point[1]}"

    # Request parameters
    params = {
        'geometries': 'geojson',
        'access_token': mapbox_access_token
    }

    # Send GET request to Mapbox Directions API
    response = requests.get(url, params=params)

    # Get the JSON response body
    route = response.json()
    print(route)
    # Check if there is at least one route
    # if 'routes' in route and len(route['routes']) > 0:
        # Create a LineString from the route coordinates and add it to the list of routes
        # line = LineString(route['routes'][0]['geometry']['coordinates'])
    allroutes.append(route['routes'])
```

```python
#This is a list of list of dictionaries
allroutes
```

```python
#creating a list of dictionaries
geometry = []
```

```python
for sublist in allroutes:
    for dictionary in sublist:
        geometry.append(dictionary['geometry'])
```

```python
geometry
```

```python
# Create a list of LineString geometries
geometries = []
for entry in geometry:
    coordinates = entry['coordinates']
    line = LineString(coordinates)
    geometries.append(line)
```

```python
# Create a GeoDataFrame with the LineString geometries
gdf = gpd.GeoDataFrame(geometry=geometries)
```

```python
# Output GeoDataFrame to GeoJSON file
output_file = 'output.geojson'
gdf.to_file(output_file, driver='GeoJSON')
```
